```python
#Age Distribution
sns.histplot(
    data=df,
    x='age',
    kde=True
)

plt.title("Age Distribution")
plt.show()
```

```python
#Daily Screen Time Distribution
sns.histplot(
    data=df,
    x='daily_screen_time_hours',
    kde=True
)

plt.title("Daily Screen Time")
plt.show()
```

```python
#Weekend Screen Time
sns.histplot(
    data=df,
    x='weekend_screen_time',
    kde=True
)

plt.title("Weekend Screen Time")
plt.show()
```

```python
#Sleep Distribution
sns.histplot(
    data=df,
    x='sleep_hours',
    kde=True
)

plt.title("Sleep Hours")
plt.show()
```

```python
#Screen Time vs Sleep
sns.scatterplot(
    data=df,
    x='daily_screen_time_hours',
    y='sleep_hours'
)

plt.title("Screen Time vs Sleep")
plt.show()
```

```python
numeric_df = df.select_dtypes(
    include=['number']
)

corr = numeric_df.corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
corr_pairs = (
    corr.unstack()
    .sort_values(
        ascending=False
    )
)

corr_pairs.head(30)
```

Research Findings

1. Screen time negatively correlates with sleep.

2. Users with higher addiction levels spend more time on social media.

3. Notification count positively correlates with app opens.

4. High gaming usage is associated with poorer academic impact.

5. Weekend screen time is significantly higher than weekday usage.
