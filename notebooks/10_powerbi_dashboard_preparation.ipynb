```python
df = pd.read_csv(
    "dashboard_dataset.csv"
)
```

```python
df.head()
df.shape
df.columns.tolist()
df.info()
```

```python
#Generate KPIs
#Average DEI

avg_dei = round(
    df['DEI'].mean(),
    2
)

print(avg_dei)

#Average NDI

avg_ndi = round(
    df['NDI'].mean(),
    2
)

print(avg_ndi)

#Average Attention

avg_attention = round(
    df['attention_score'].mean(),
    2
)

print(avg_attention)

#Average Aggression

avg_aggression = round(
    df['Aggression_Risk'].mean(),
    2
)

print(avg_aggression)

#Average Wellbeing

avg_wellbeing = round(
    df['Wellbeing_Score'].mean(),
    2
)

print(avg_wellbeing)
```

```python
#KPI Summary Table

kpi_table = pd.DataFrame({

    'Metric': [

        'Average DEI',

        'Average NDI',

        'Average Attention',

        'Average Aggression',

        'Average Wellbeing'

    ],

    'Value': [

        avg_dei,

        avg_ndi,

        avg_attention,

        avg_aggression,

        avg_wellbeing

    ]

})

kpi_table

```

```python
kpi_table.to_csv(
    "kpi_summary.csv",
    index=False
)
```
